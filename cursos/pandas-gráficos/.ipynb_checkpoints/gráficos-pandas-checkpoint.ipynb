```python
import pandas as pd
```

<h2>Leer CSV con Pandas</h2>

<p>
    La función <code>pd.read_csv</code> permite leer un archivo CSV y permite definir distintos parámetros. Para este curso resaltaremos 3:
    <ol>
    <li><strong>ruta_archivo:</strong> Este es el primer parámetro que se entrega en la función. Corresponde a un string con la ruta en la que se encuentre el archivo. Puede ser la ruta completa (<code>C:/User/nombre/carpeta/archivo.csv</code> o la dirección a un archivo que se encuentre dentro de la misma carpeta del código (<code>./carpeta/archivo.csv</code>).</li>
    <li><strong>sep:</strong> Se entrega un carácter (letra, número, simbolo, etc) que define la separación de un campo con otro, tanto entre los nombres de campos como los valores. Este parámetro puede ser omitido para que use por defecto la ",", pero igualmente se entrega en este ejemplo para mostrar como funciona.</li>
    <li><strong>encoding:</strong> Formato en que se leerá el texto del archivo. En caso de archivos con carácteres que posean tildes o "ñ" se recomienda utilizar "latin1", de lo contrario se puede omitir.</li>
    </ol>
    Habiendo entendido como utilizar la función, ya pdoemos trabajar con ella para cargar la descripción de las películas e imprimir la tabla con el comando <code>display()</code>.
</p>

```python
# Se utiliza la función de "pandas.read_csv()" para leer un archivo de texto (o CSV) y cargarlo en un DataFrame.
movie_description = pd.read_csv(
    "./data/movie_description.csv", # representa a la ruta en la que se encuentra el archivo.
    sep=",", # indica el carácter que utiliza como separador el archivo, por defecto siempre utiliza ","
    encoding="latin-1" # utilizado para para identificar el formato del archivo. Se recomienda usar "latin-1" cuando haya texto con tildes o "ñ".
)

# Imprimir DataFrame recién creado con la descripción de la película
display(movie_description)
```

<p>
    Una vez revisada y entendida la estructura de la tabla de películas, podemos cargar los datos para comenzar a trabajar y se puede utilizar la misma función anterior, solo que esta vez la guardaremos en una nueva variable llamada <code>movie</code>.
</p>

```python
# En este caso omitimos el parámetro de encoding, dado que toda la información es de una fuente en inglés
movie = pd.read_csv("./data/movie.csv")

# Al imprimir solo se muestra una porción de la tabla, al tener varios registros (4916), existen funciones para ampliar el rango
display(movie)
```

```python
with pd.option_context('display.max_rows', 20, 'display.max_columns', 30):
    display(movie)
```

<h2>Consultando la tabla</h2>
<p>
    Aplicaremos técnicas que ya hemos aprendido en el curso anterior y haremos pequeñas consultas a la tabla de películas, usando distintos filtros. Para esto se irán proponiendo una serie de problemas para ir cumpliendo con cada requerimiento.
</p>
<h3>Ejemplo 1</h3>
<p>
    Se requiere conocer los títulos de películas hechas en Estados Unidos (USA) con sus respectivos directores y solo si se conoce la información de este último.
</p>

```python
movie_ej1 = movie[
    (movie["country"] == "USA") & ~(movie["director_name"].isnull())
][
    ["movie_title", "director_name"]
]
display(movie_ej1)
```

<p>
    En este caso se requieren hacer dos filtros: uno para traer solo aquellas películas que sean de Estados Unidos y aquellas en los que se tenga el nombre del direction (que este no sea nulo). Para lograr esto requerimos de agregar ambas condiciones conectadas por "&" quien representa el conector lógico de "y". Por último, ya que solo se requieren los campos del título de la película (movie_title) y el nombre del director (director_name) se entrega estos dos valores como lista después del filtro. Se sigue la lógica de:
</p>

<center><code>movie[(Condición 1) & (Condición 2)][(Lista de campos)]</code></center>

<h3>Ejemplo 2</h3>
<p>
    Ahora se requiere tener un listado de todas aquellas películas con clasificación "PG-13" y que hayan salido entre el año 2000 y 2005.
</p>

```python
movie_ej2 = movie[(movie["title_year"] >= 2000) & (movie["title_year"] <= 2005) & (movie["content_rating"] == "PG-13")][["movie_title"]]
display(movie_ej2)
```

<p>
    En este caso se requerían de tres condiciones: una para las películas que salieron después del 2000 y antes del 2005 (se debe escribir una por cada año) y para filtrar la clasificación de "PG-13". Como unicamente se requieren las películas, solo se extrae este campo.
</p>

<h2>Operaciones sobre campos</h2>
<p>
    Además de poder realizar filtros, también podemos crear nuevos campos a partir del cálculo de otros o usando simplemente valores constantes. En este caso, y a modo de ejemplo, tomaremos las ganancias (gross) que obtuvo la película y obtendremos cual fue la variación con respecto al presupuesto (budget) que esta tenía. El resultado será almacenado en un nuevo campo.
    La formula corresponde a la siguiente: $$Var = \frac{Gross-Budget}{Budget}$$
</p>

```python
movie["var"] = (movie["gross"]-movie["budget"])/movie["budget"]
display(movie[["movie_title", "gross", "budget", "var"]])
```

<p>
    Se observa que los campos pueden ser trabajados usando operaciones matemáticas sin hacer muchos cambios, solo identificando cada uno de ellos por el DataFrame y su nombre y se puede tanto imprimir como asignar a un nuevo campo (como se hizo en este caso).
</p>

<h2>Cálculos agrupados</h2>
<p>
    Pandas permite realizar cálculos agrupados, es decir, de acuerdo a una parte de campos de la tablas, se puede hacer distintas operaciones sobre otros campos y agrupar los resultados para reducir la tabla. Por ejemplo, si quisieramos conocer el número de likes total que ha tenido un director de acuerdo a las películas en las que trabajó, sería de la siguiente forma:
</p>

```python
likes_director = movie.groupby("director_name", as_index=False)["movie_facebook_likes"].sum()
display(likes_director)
```

<p>
    La función <code>groupby()</code> indica de acuerdo a que campos se desea agrupar la información y el parámetro <code>as_index</code>, que toma un valor de tipo booleano, permite definir si se quiere recibir una lista con indices que contenga los resultados (<code>True</code>) o que devuelva un DataFrame con un formato similar (<code>False</code>).
</p>

<h3>Agrupar por más de un campo</h3>
<p>
    Igualmente se pueden definir más de un campo para agrupar. Continuando con el ejemplo anterior, supongamos que ahora deseamos agrupar también por clasificación de la película (content_rating). Para lograrlo debería ser de la siguiente forma:
</p>

```python
rating_director = movie.groupby(["director_name", "content_rating"], as_index=False)["movie_facebook_likes"].sum()
display(rating_director)
```

<p>
    La única diferencia con la función anterior es que en lugar de entregar solo un string (con el nombre del campo) se pasa una lista de string, pudiendo agrupa por más de un solo campo.
</p>
<h3>Hacer más de un cálculo</h3>
<p>
    Así como se puede agrupar por más de un campo, también podemos realizar más de un cálculo en una sola operación. Para esto es importante apoyarnos en otra biblioteca llamada numpy. Se importa de forma similar como se hizo con pandas. Asumiremos que, además de los likes, ahora también queremos tener la variación promedio, el campo que acabamos de cálculas. Para lograr esto y, una vez importado el paquete de numpy, se realiza la siguiente operación:
</p>

```python
import numpy as np
```

```python
likes_var = movie.groupby(["director_name", "content_rating"], as_index=False).agg(
    movie_facebook_likes=("movie_facebook_likes", np.sum),
    var=("var", np.mean)
)
display(likes_var)
```

<p>
    Se cambia por completo la parte de agregación, ahora incluyendo la función <code>agg()</code> en el que se define los campos que se deseen calcular, el nombre que estos tendrán como resultado y el tipo de operación a realizar, en este último punto es donde se apoya en la biblioteca de <code>numpy</code>.
</p>

<h2>Graficando con Pandas</h2>
<p>
    Por último, podemos crear gráficos a partir de nuestros datos, estos pueden ser de barra, apilados, torta, puntos, etc. Antes de poder entrar directamente a a gráfica, debemos importar una nueva biblioteca, especificamente la de <code>matplot</code>.
</p>

```python
import matplotlib.pyplot as plt
```

<h3>Gráfico de barras</h3>
<p>
    Para nuestro primer caso crearemos un gráfico de puntos. Para esto se requiere ejecutar las siguientes líneas de código.
</p>

```python
graph1 = movie.groupby("content_rating", as_index=False)["movie_facebook_likes"].sum()
display(graph1)
```

```python
## Gráfico de barras

graph1.plot(
    kind="bar",
    x="content_rating", # Eje horizontal
    y ="movie_facebook_likes", # Eje vertical
    color="#000000", # Color del gráfico
    figsize=(12,12)
)
```

```python
## Gráfico de puntos

graph1.plot(
    kind="scatter",
    x="content_rating", # Eje horizontal
    y ="movie_facebook_likes", # Eje vertical
    color="purple" # Color del gráfico
)
```

```python
## Gráfico de barras por grupos

graph2 = movie.groupby(["content_rating", "title_year"], as_index=False)["movie_facebook_likes"].sum()
graph2 = graph2[graph2["title_year"].isin([2010, 2011, 2012])][["content_rating", "title_year", "movie_facebook_likes"]]
display(graph2)
```
